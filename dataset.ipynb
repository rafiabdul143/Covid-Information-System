```python
import tkinter as tk
from tkinter import ttk
import pandas as pd

# Define your dataset
data = {
    "Country/Region": (
        ["India"] * 30 + ["United States"] * 30 + ["Brazil"] * 30 + ["Russia"] * 30 + ["China"] * 30 +
        ["Germany"] * 30 + ["France"] * 30 + ["UK"] * 30 + ["Canada"] * 30 + ["Australia"] * 30
    ),
    "Date": (
        ["2022-01-01", "2023-01-01", "2024-01-01"] * 100
    ),
    "Confirmed": (
        [100000, 200000, 300000, 150000, 250000, 400000, 80000, 150000, 300000, 70000, 130000, 220000,
         50000, 100000, 180000, 90000, 160000, 260000, 85000, 140000, 250000, 60000, 120000, 230000,
         95000, 180000, 320000, 90000, 160000, 290000] * 10
    )
}

dataset = pd.DataFrame(data)

# Create the Tkinter window
root = tk.Tk()
root.title("COVID-19 Dataset")
root.geometry("900x700")
root.configure(bg="black")

# Title Label
title_label = tk.Label(
    root, text="COVID-19 Dataset", font=("Helvetica", 24, "bold"), bg="black", fg="red", pady=20
)
title_label.pack(fill=tk.X)

# Create the Treeview for displaying the dataset
tree_frame = tk.Frame(root, bg="black")
tree_frame.pack(fill=tk.BOTH, expand=True)

tree = ttk.Treeview(tree_frame, columns=("Country/Region", "Date", "Confirmed"), show="headings", height=15)

# Style the Treeview
style = ttk.Style()
style.configure("Treeview",
                background="black",
                foreground="red",
                rowheight=25,
                font=("Arial", 12),
                fieldbackground="black")
style.configure("Treeview.Heading",
                background="red",
                foreground="black",
                font=("Arial", 14, "bold"),
                relief="raised")  # Adds a raised effect to the headings

# Define the column headings
tree.heading("Country/Region", text="Country/Region")
tree.heading("Date", text="Date")
tree.heading("Confirmed", text="Confirmed Cases")

# Define the column widths
tree.column("Country/Region", width=200, anchor="center")
tree.column("Date", width=150, anchor="center")
tree.column("Confirmed", width=150, anchor="center")

# Insert the data into the Treeview
for _, row in dataset.iterrows():
    tree.insert("", "end", values=(row["Country/Region"], row["Date"], row["Confirmed"]))

tree.pack(fill=tk.BOTH, expand=True, side="left")

# Scrollbars for the Treeview
y_scroll = ttk.Scrollbar(tree_frame, orient="vertical", command=tree.yview)
y_scroll.pack(side="right", fill="y")

tree.configure(yscrollcommand=y_scroll.set)

# Exit Button
exit_button = tk.Button(
    root, text="Exit", command=root.quit, bg="red", fg="black", font=("Arial", 12), padx=20, pady=10
)
exit_button.pack(pady=20)

# Run the Tkinter event loop
root.mainloop()
```
