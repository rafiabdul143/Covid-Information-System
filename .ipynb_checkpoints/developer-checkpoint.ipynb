```python
from tkinter import *
from PIL import Image, ImageTk, ImageFilter
import webbrowser  # Ensure webbrowser module is imported


class Developer:
    def __init__(self, root):
        self.root = root
        self.root.geometry("1530x790+0+0")
        self.root.title("Covid Information Center")

        # Title Label
        title_lb1 = Label(self.root, text="DEVELOPERS INFO", font=("times new roman", 35, "bold"), bg="black", fg="red")
        title_lb1.place(x=0, y=0, width=1530, height=45)

        # Background Image
        img3 = Image.open(r"3771873.jpg")
        img3 = img3.resize((1530, 790), Image.LANCZOS)
        img3 = img3.filter(ImageFilter.GaussianBlur(radius=5))
        self.photoimg3 = ImageTk.PhotoImage(img3)

        bg_img = Label(self.root, image=self.photoimg3)
        bg_img.place(x=0, y=45, width=1530, height=790)

        # Add a styled text label
        self.intro_label = Label(
            self.root,
            text="",
            font=("Helvetica", 14, "italic"),
            bg="black",
            fg="#FF0000",
            wraplength=1000,
            justify="left",
            padx=20,
            pady=20,
        )
        self.intro_label.place(x=250, y=80, width=1000, height=420)

        # Add typing effect
        self.add_intro_text()

        # Add "Developed by" section
        self.add_developer_names()

    def add_intro_text(self):
        # Styled text
        self.text = (
            "As a Computer Science students, We developed this project as part of our coursework. \n"
        "It prominently features a fully functioning GUI, developed using Tkinter, making the "
        "application intuitive and user-friendly.\n"
            "Various Python libraries were utilized to enhance its functionality and design:\n"
            "- Tkinter: For creating the graphical user interface (GUI).\n"
            "- PIL (Pillow): For image processing tasks.\n"
            "- Matplotlib (optional): For future additions like graphical data representation.\n"
            "- ImageFilter: For applying visual effects, such as Gaussian blur, to enhance the aesthetics.\n"
            "- Pandas: For efficient data manipulation and analysis.\n"
            "- Seaborn: For creating advanced and aesthetically pleasing data visualizations.\n"
            "We invite you to follow me on the provided social media handles. I would be delighted to hear any "
            "modifications or innovative ideas you might have for this project, and I will definitely consider "
            "your suggestions.\n\nThank you!"
        )

        self.index = 0  # Starting index for the text
        self.typing_effect()  # Start the typing effect

    def typing_effect(self):
        if self.index < len(self.text):
            # Add one character at a time to the label text
            self.intro_label.config(text=self.text[: self.index + 1])
            self.index += 1
            # Call this method again after a delay
            self.root.after(20, self.typing_effect)  # Adjust delay to change typing speed

    def add_developer_names(self):
        # Developer names label with LinkedIn logo beside each name
        developer_label = Label(
            self.root,
            text="Developed By:",
            font=("Helvetica", 18, "bold"),
            bg="black",
            fg="#39FF14",
            justify="center",
        )
        developer_label.place(x=600, y=530, width=400, height=50)

        # Developer details
        self.add_developer_with_logo(
            "Abdul Rafi",
            linkedin_url="https://www.linkedin.com/in/abdulrafi0870",
            github_url="https://github.com/rafiabdul143",
            gmail_url="mailto:rafiabdul143@gmail.com",
            x_start=600,
            y_start=600,
        )
        self.add_developer_with_logo(
            "Nishath Sultana",
            linkedin_url="https://www.linkedin.com/in/nishath-sultana-86233b251",
            github_url="https://github.com/nishath742",
            gmail_url="mailto:nishathsultana742@gmail.com",
            x_start=600,
            y_start=650,
        )

    def add_developer_with_logo(self, name, linkedin_url, github_url, gmail_url, x_start, y_start):
        # Create a frame to contain the name and the logos
        frame = Frame(self.root, bg="black")
        frame.place(x=x_start, y=y_start, width=400, height=50)

        # Developer name label inside the frame
        name_label = Label(frame, text=name, font=("Helvetica", 14), bg="black", fg="#39FF14")
        name_label.pack(side=LEFT, padx=10)  # Adding some padding between name and logos

        # LinkedIn icon
        linkedin_icon_path = r"images\linkedIn_PNG8.png"
        linkedin_icon_size = (30, 30)  # Resize to fit beside name
        linkedin_img = Image.open(linkedin_icon_path).resize(linkedin_icon_size, Image.LANCZOS)
        linkedin_photo = ImageTk.PhotoImage(linkedin_img)

        linkedin_btn = Button(
            frame, image=linkedin_photo, command=lambda link=linkedin_url: webbrowser.open(link), bd=0, bg="black"
        )
        linkedin_btn.image = linkedin_photo  # Keep a reference to avoid garbage collection
        linkedin_btn.pack(side=LEFT, padx=5)  # Place beside the name label

        # GitHub icon
        github_icon_path = r"images\gitlog.jpg"
        github_icon_size = (30, 30)  # Resize to fit beside LinkedIn
        github_img = Image.open(github_icon_path).resize(github_icon_size, Image.LANCZOS)
        github_photo = ImageTk.PhotoImage(github_img)

        github_btn = Button(
            frame, image=github_photo, command=lambda link=github_url: webbrowser.open(link), bd=0, bg="black"
        )
        github_btn.image = github_photo  # Keep a reference to avoid garbage collection
        github_btn.pack(side=LEFT, padx=5)  # Place beside the LinkedIn button

        # Gmail icon
        gmail_icon_path = r"images\Gmail-Logo.png"
        gmail_icon_size = (30, 30)  # Resize to fit beside GitHub
        gmail_img = Image.open(gmail_icon_path).resize(gmail_icon_size, Image.LANCZOS)
        gmail_photo = ImageTk.PhotoImage(gmail_img)

        gmail_btn = Button(
            frame, image=gmail_photo, command=lambda link=gmail_url: webbrowser.open(link), bd=0, bg="black"
        )
        gmail_btn.image = gmail_photo  # Keep a reference to avoid garbage collection
        gmail_btn.pack(side=LEFT, padx=5)  # Place beside the GitHub button


if __name__ == "__main__":
    root = Tk()
    app = Developer(root)
    root.mainloop()
```
