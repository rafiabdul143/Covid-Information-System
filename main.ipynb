```python
from tkinter import *
from PIL import Image, ImageTk, ImageFilter  # Added ImageFilter import
import nbimporter
from developer import Developer
from helpdesk import Helpdesk
import subprocess  # Import subprocess

class CovidInformationSystem:
    def __init__(self, root):
        self.root = root
        self.root.geometry("1530x790+0+0")
        self.root.title("Covid Information Center")

        # Background image with Gaussian Blur
        try:
            img3 = Image.open(r"images\bg1.jpg")
            img3 = img3.resize((1530, 710), Image.LANCZOS)

            # Apply Gaussian Blur
            img3 = img3.filter(ImageFilter.GaussianBlur(radius=10))  # Adjust the radius for more or less blur
            self.photoimg3 = ImageTk.PhotoImage(img3)  # Store the image as an instance variable

            bg_img = Label(self.root, image=self.photoimg3)
            bg_img.place(x=0, y=0, width=1530, height=790)
        except FileNotFoundError:
            print("Background image not found. Please check the path.")

        # Title label
        title_lb1 = Label(
            self.root,
            text="COVID INFORMATION CENTER",
            font=("times new roman", 35, "bold"),
            bg="black",
            fg="red",
        )
        title_lb1.place(x=0, y=0, width=1530, height=130)

        # Additional text above buttons
        text_line1 = Label(
            self.root,
            text='"Every Statistics Represents a Story; \nEvery Life Lost is a Tragedy"',
            font=("Gabriola", 20, "italic"),  # Stylish font with italic style
            bg="black",
            fg="white",
        )
        text_line1.place(x=500, y=200, width=550, height=80)

        # Center coordinates for the square layout
        center_x = 765  # Half of 1530 (screen width)
        center_y = 400  # Approx middle of 790 (screen height)
        button_width = 220
        button_height = 40
        button_gap = 50  # Gap between buttons

        # Button 1: Top-left
        b1 = Button(
            self.root,
            text="Visualize Data",
            command=self.statistcs,
            cursor="hand2",
            font=("times new roman", 15, "bold"),
            bg="#333333",
            fg="#39FF14",
        )
        b1.place(
            x=center_x - button_width - button_gap // 2,
            y=center_y - button_height - button_gap // 2,
            width=button_width,
            height=button_height,
        )

        # Button 2: Top-right
        b2 = Button(
            self.root,
            text="View Dataset",
            command=self.Socho,
            cursor="hand2",
            font=("times new roman", 15, "bold"),
            bg="#333333",
            fg="#39FF14",
        )
        b2.place(
            x=center_x + button_gap // 2,
            y=center_y - button_height - button_gap // 2,
            width=button_width,
            height=button_height,
        )

        # Button 3: Bottom-left
        b3 = Button(
            self.root,
            text="Helpdesk",
            command=self.helpdesk,
            cursor="hand2",
            font=("times new roman", 15, "bold"),
            bg="#333333",
            fg="#39FF14",
        )
        b3.place(
            x=center_x - button_width - button_gap // 2,
            y=center_y + button_gap // 2,
            width=button_width,
            height=button_height,
        )

        # Button 4: Bottom-right
        developer = Button(
            self.root,
            text="Developer",
            command=self.developer,
            cursor="hand2",
            font=("times new roman", 15, "bold"),
            bg="#333333",
            fg="#39FF14",
        )
        developer.place(
            x=center_x + button_gap // 2,
            y=center_y + button_gap // 2,
            width=button_width,
            height=button_height,
        )

        # Close button at the bottom
        btn_close = Button(
            self.root,
            text="Close Application",
            command=self.close_app,
            font=("times new roman", 15, "bold"),
            bg="red",
            fg="white",
            cursor="hand2",
        )
        btn_close.place(
            x=center_x - button_width // 2,
            y=center_y + button_height + 2 * button_gap,
            width=button_width,
            height=button_height,
        )

    def statistcs(self):
        try:
            # Running the Jupyter notebook using subprocess
            subprocess.run(
                ["jupyter", "nbconvert", "--execute", "statistcs.ipynb", "--to", "notebook", "--inplace"],
                check=True,
            )
            print("Success", "Notebook executed successfully!")
        except subprocess.CalledProcessError as e:
            print("Error", f"Error executing notebook: {e}")

    def Socho(self):
        try:
            # Running the Jupyter notebook using subprocess
            subprocess.run(
                ["jupyter", "nbconvert", "--execute", "dataset.ipynb", "--to", "notebook", "--inplace"],
                check=True,
            )
            print("Success", "Notebook executed successfully!")
        except subprocess.CalledProcessError as e:
            print("Error", f"Error executing notebook: {e}")

    def helpdesk(self):
        self.new_window = Toplevel(self.root)
        self.app = Helpdesk(self.new_window)

    def developer(self):
        try:
            # Running the Jupyter notebook using subprocess
            subprocess.run(
                ["jupyter", "nbconvert", "--execute", "developer.ipynb", "--to", "notebook", "--inplace"],
                check=True,
            )
            print("Success", "Notebook executed successfully!")
        except subprocess.CalledProcessError as e:
            print("Error", f"Error executing notebook: {e}")

    def close_app(self):
        self.root.destroy()


if __name__ == "__main__":
    root = Tk()
    app = CovidInformationSystem(root)
    root.mainloop()
```
